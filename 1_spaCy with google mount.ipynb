```python
# Import SpaCy library
import spacy

# Instantiate NLP pipeline
nlp = spacy.load('en_core_web_sm')

# Assign the filename to a variable
# filename = '/content/drive/My Drive/1_Study-Projects/Colab Notebooks/EnDyna_Demo_SampleData.txt'

# Make the text of the file available to our script
# ner_text = open(filename).read()
ner_text = 'Threat actors exploiting Log4Shell (CVE-2021-44228) and CVE-2021-45046.\n C2 server observed at 185.220.101.1 with domain bad-malware[.]com.\n Associated file hash: d41d8cd98f00b204e9800998ecf8427e'

# Create the Doc object by passing it through the text pipeline (nlp)
doc = nlp(ner_text)

# Run NER on the Doc object
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_, spacy.explain(ent.label_))
```
